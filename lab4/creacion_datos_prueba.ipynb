```python
seed = 161
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = 'example_data/202220_Laboratorio_3_data_university_admission_train.csv'
df = pd.read_csv(ruta)
print("Número de filas: ", df.shape[0])
print("Número de columnas: ", df.shape[1])
df.sample(5)
```

```python
# Delete row Unnamed: 0
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Rename columns
df.rename(columns={'Serial No.': 'serial_no','GRE Score': 'gre_score', 'TOEFL Score': 'toefl_score', 'University Rating': 'university_rating', 'SOP': 'sop', 'LOR ': 'lor', 'CGPA': 'cgpa', 'Research': 'research'}, inplace=True)
```

```python
import json

json_array = []
# For each row in df convert to JSON
for i, row in df.iterrows():
    json_array.append(row.to_json())

# Convert dataframe to json
json_array = json.loads(df.to_json(orient='records'))

# Save JSON array to file
with open('./datos_prueba/caso_prueba5.json', 'w') as f:
    f.write(json.dumps(json_array))
```

```python
# Get columns names
columns = df.columns.tolist()

random_data = []
for i in range(5):
    row = {}
    for col in columns:
        if col == 'serial_no':
            row[col] = np.random.randint(1, 1000)
        elif col == 'research':
            row[col] = np.random.randint(0, 1)
        elif col == 'university_rating':
            row[col] = np.random.randint(1, 5)
        elif col == 'gre_score':
            row[col] = np.random.randint(0, 340)
        elif col == 'toefl_score':
            row[col] = np.random.randint(0, 120)
        elif col == 'sop':
            row[col] = np.random.uniform(0, 5)
        elif col == 'lor':
            row[col] = np.random.uniform(0, 5)
        elif col == 'cgpa':
            row[col] = np.random.uniform(0, 10)
    random_data.append(row)


# Save JSON array to file
with open('./datos_prueba/caso_prueba2.json', 'w') as f:
    f.write(json.dumps(random_data))
```
